# Classes of colormaps
In Matplotlib, color maps (colormaps) are classified into two main categories:

Sequential Colormaps:
Sequential colormaps consist of a continuous sequence of colors, ranging from low to high intensity. They are best suited for representing data that has a natural ordering or progression. For example, you might use a sequential colormap to visualize temperature variations, where colors transition smoothly from cold to hot.

Diverging Colormaps:
Diverging colormaps are designed for data with a central critical value or pivot point. The colors diverge from this central value, representing data that can have both positive and negative deviations from the midpoint. Diverging colormaps are commonly used when displaying data with a meaningful center, such as anomalies relative to a baseline value.

Matplotlib offers a variety of built-in colormaps in each class, and you can choose the most suitable colormap based on the nature of your data and the visual representation you want to achieve.

there are also,
Cyclic: change in lightness of two different colors that meet in the middle and beginning/end at an unsaturated color; should be used for values that wrap around at the endpoints, such as phase angle, wind direction, or time of day.

Qualitative: often are miscellaneous colors; should be used to represent information which does not have ordering or relationships.

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
schools=pd.read_csv('datasets\schools.csv')
district=gpd.read_file("datasets\school_districts.geojson")
```

```python
schools.head()
```

```python
district.plot(column = 'district', legend = True, cmap = 'Set2')
plt.scatter(schools.Longitude, schools.Latitude, color='b', marker=2)
# Add labels and title
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.title('Nashville Schools and School Districts')
plt.grid()
plt.show(); 
```

# Colormaps
When your intention is to distinguish between regions without suggesting any specific relationship between them, opting for a qualitative colormap is the most suitable approach.
